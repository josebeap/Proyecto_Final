```python
from google_images_search import GoogleImagesSearch
from threading import Thread
import queue
import requests
from PIL import Image
from io import BytesIO

api_key = 'AIzaSyBdC_ol-yZGKMf4kaaQKioImBRChTJMh0w'
cx = 'f17ae98bd36e043fa'
gis = GoogleImagesSearch(api_key, cx)

def cargar_imagenes(q, imagenes):
    while True:
        try:
            search_params = q.get(timeout=3)  # Espera 3 segundos o hasta que haya un elemento en la cola
        except queue.Empty:
            break  # Si no hay elementos durante 3 segundos, el hilo termina

        # Realiza la búsqueda
        gis.search(search_params=search_params)
        for image in gis.results():
            # Obtiene la URL de la imagen y la carga en memoria
            response = requests.get(image.url)
            img = Image.open(BytesIO(response.content))
            imagenes.append(img)

# Cola para almacenar los parámetros de búsqueda
search_queue = queue.Queue()

# Lista para almacenar las imágenes cargadas
imagenes_en_memoria = []

# Cargar la cola con los parámetros de búsqueda
for i in range(50):
    search_params = {
        'q': 'término de búsqueda',  # Modifica esto según lo que quieras buscar
        'num': 1
    }
    search_queue.put(search_params)

# Crear y comenzar los hilos
threads = []
for i in range(10):
    thread = Thread(target=cargar_imagenes, args=(search_queue, imagenes_en_memoria))
    thread.start()
    threads.append(thread)

# Esperar a que todos los hilos terminen
for thread in threads:
    thread.join()

print(f"Se han cargado {len(imagenes_en_memoria)} imágenes en memoria.")
```
